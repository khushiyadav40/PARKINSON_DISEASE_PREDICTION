**DATA VISUALIZATION**

Target Variable Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="status", data=data)
plt.title("Distribution of Parkinson's Disease Status")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks([0,1], ["Healthy (0)", "Parkinson's (1)"])
plt.show()
```

**The dataset contains 147 Parkinson's and 48 healthy samples, indicating a class imbalance with Parkinson's cases as the majority class.**

Correlation Heatmap

```python
plt.figure(figsize=(16,12))
correlation = data.drop("name", axis=1).corr()
sns.heatmap(correlation,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Heatmap of Voice Features")
plt.show()
```

**The heatmap shows that several voice features are strongly correlated with one another, indicating possible multicollinearity. The status variable also exhibits noticeable correlations with several acoustic features, suggesting that these measurements may be useful for predicting Parkinson's disease.**

Distribution of Numerical Features

```python
data.drop(columns=["name", "status"]).hist(figsize=(18, 16), bins=15)
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

**The histograms indicate that most numerical features have right-skewed distributions, while a few features are approximately normally distributed. The varying distributions and wide spread of values suggest that feature scaling may be beneficial before model training.**

Boxplots of Important Features by Target Variable

```python
features = ["MDVP:Fo(Hz)", "MDVP:Jitter(%)", "MDVP:Shimmer", "HNR"]
plt.figure(figsize=(15,10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x="status", y=feature, data=data)
    plt.title(f"{feature} vs Status")
    plt.xlabel("Status")
    plt.xticks([0, 1], ["Healthy", "Parkinson's"])
plt.tight_layout()
plt.show()
```

**The boxplots show noticeable differences in voice characteristics between healthy individuals and Parkinson's patients. Parkinson's samples generally exhibit higher jitter and shimmer values and lower HNR values, indicating greater voice instability. These differences suggest that these features are useful for distinguishing between the two classes.**

Feature Correlation with Target Variable

```python
plt.figure(figsize=(8,10))
target_corr = data.drop(columns=["name"]).corr()["status"].drop("status").sort_values()
target_corr.plot(kind="barh")
plt.title("Correlation of Features with Target Variable")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.show()
```

**The bar plot shows that spread1, PPE, and spread2 have the strongest positive correlation with the target variable, while MDVP:Fo(Hz), MDVP:Flo(Hz), and HNR exhibit the strongest negative correlation. These features are likely to be the most influential in predicting Parkinson's disease.**
